```python
# =========================
# STEP 1: Install
# =========================
!pip install -q langchain-community langchain-huggingface sentence-transformers faiss-cpu transformers biopython requests==2.32.4

# =========================
# STEP 2: Setup
# =========================
import os
os.environ["TOKENIZERS_PARALLELISM"] = "false"

from Bio import Entrez
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from transformers import pipeline

# 👉 REQUIRED: add your email (NCBI requirement)
Entrez.email = "your_email@example.com"

# =========================
# STEP 3: Fetch PubMed Abstracts
# =========================
def fetch_pubmed_abstracts(query, max_results=5):
    search = Entrez.esearch(db="pubmed", term=query, retmax=max_results)
    record = Entrez.read(search)
    ids = record["IdList"]

    fetch = Entrez.efetch(db="pubmed", id=",".join(ids), rettype="abstract", retmode="text")
    abstracts = fetch.read().split("\n\n")

    # Clean abstracts
    docs = [a.strip() for a in abstracts if len(a.strip()) > 50]
    return docs[:max_results]

# =========================
# STEP 4: Load Embeddings + LLM
# =========================
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

generator = pipeline(
    "text-generation",
    model="google/flan-t5-base",
    max_new_tokens=120
)

# =========================
# STEP 5: RAG Pipeline Function
# =========================
def rag_pubmed_agent(query):
    print("\n🔎 Fetching abstracts from PubMed...")

    docs = fetch_pubmed_abstracts(query, max_results=5)

    if not docs:
        print("No abstracts found.")
        return

    # Create vector DB
    db = FAISS.from_texts(docs, embedding)
    retriever = db.as_retriever(search_kwargs={"k": 2})

    # Retrieve relevant context
    retrieved_docs = retriever.invoke(query)
    context = " ".join([d.page_content for d in retrieved_docs])

    # Prompt
    prompt = f"""
    Answer based on the biomedical abstracts:

    Context: {context}

    Question: {query}

    Answer:
    """

    # Generate answer
    result = generator(prompt)[0]['generated_text']

    # Output
    print("\n📄 Top Retrieved Abstracts:")
    for i, d in enumerate(retrieved_docs, 1):
        print(f"\n{i}. {d.page_content[:300]}...")

    print("\n🤖 Answer:")
    print(result)
    print("="*60)

# =========================
# STEP 6: Interactive Query
# =========================
print("🔁 PubMed RAG Agent Ready (type 'exit' to stop)")

while True:
    user_query = input("\nEnter your biomedical query: ")

    if user_query.lower() == "exit":
        print("Exiting...")
        break

    rag_pubmed_agent(user_query)
```
